```python
ips = [f'10.10.100.{i}' for i in range(1, 255)]
```

Guided by [PEP 3101](https://peps.python.org/pep-3101/) align the lines and use the `end='' if (i + 1) % 3 else '\n'` argument to divide the output into three lines for easier reading

```python
for i, ip in enumerate(ips):
    print(f"{i:>3} -> {ip:<16}", end='' if (i + 1) % 3 else '\n')
```

```python
for ip in ips[2:6]:
    print(ip)
```
